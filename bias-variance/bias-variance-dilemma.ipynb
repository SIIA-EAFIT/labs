Hagamos un peque√±o experimento, generemos multiples datasets de entrenamiento $\mathbb{D}_1, \mathbb{D}_2, \dots, \mathbb{D}_M$ y con el entrenemos multiples modelos $f_1, f_2, \dots, f_M$ y observemos como distribucionalmente dichos modelos aproximan la funcion que nos interesa.

```python
import matplotlib.pyplot as plt
import torch

from src.poly import PolinomialFunction, PolynomialRegression
from src.bias_variance import bias_variance_experiment, plot_bias_variance
```

```python
torch.manual_seed(10); # just to have a nice graph

oracle = PolinomialFunction(5, noise_std=0.5)
```

```python
model = PolynomialRegression(0)

plot_bias_variance(
    *bias_variance_experiment(
        oracle=oracle, 
        model=model,
        num_train_datasets=100,
        num_train_examples_per_dataset=500
    ),
    dpi=224,
)
```

```python
model = PolynomialRegression(3)

plot_bias_variance(
    *bias_variance_experiment(
        oracle=oracle, 
        model=model,
        num_train_datasets=100,
        num_train_examples_per_dataset=500
    ),
    dpi=224
)
```

```python
model = PolynomialRegression(4)

plot_bias_variance(
    *bias_variance_experiment(
        oracle=oracle, 
        model=model,
        num_train_examples_per_dataset=500
    ),
    dpi=224
)
```
