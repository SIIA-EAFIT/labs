Definamos
$$ P(x; \alpha) = \sum_{i=0}^D \alpha_i x^i $$

Consideremos ahora una regularizacion [que incentiva la suavidad](https://en.wikipedia.org/wiki/Smoothing_spline)

$$ Q(\alpha) = \int_0^1 (P''(x))^2 dx  \\ Q(\alpha) = \sum_{q=2}^D\sum_{r=2}^D \beta_{q, r}\alpha_q\alpha_r$$

donde 
$$ \beta_{q,r} = \frac{(q-1)q(r-1)r}{q + r - 3} $$

Así optimizaremos
$$argmin_{\alpha} \sum_{n=1}^N (P(x_n; \alpha) - y_n)^2 + \rho Q(\alpha)$$
